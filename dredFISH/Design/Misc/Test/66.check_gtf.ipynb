```python
import pandas as pd
import numpy as np
```

```python
f = "/bigstore/GeneralStorage/fangming/projects/dredfish/packages/PaintSHOP_pipeline/test2/data/mm10.refGene.gtf"
df = pd.read_csv(f, sep="\t", header=None)
df = df[df[2]=='exon']
df[9]  = df[8].apply(lambda x: x.split(';')[0][len("gene_id__"):-1])
df[10] = df[8].apply(lambda x: x.split(';')[1][len(" transcript_id__"):-1])
df
```

```python
gtl = df.groupby([9,0,3,4]).first().reset_index()
gtl['len'] = gtl[4]-gtl[3]+1 # +1 or not?
gtl['num30mer'] = (gtl['len']/30).astype(int) # 
gtl
```

```python
gene_probes = gtl.groupby(9)[['len', 'num30mer']].sum()
gene_probes
```

```python
gene_probes['num30mer_cont'] = (gene_probes['len']/30).astype(int)
gene_probes.index.name = 'gene'
gene_probes
```

```python
output = "/bigstore/GeneralStorage/fangming/projects/dredfish/packages/PaintSHOP_pipeline/test2/data/num30mer_estimate.csv"
gene_probes.to_csv(output, header=True, index=True)
```
